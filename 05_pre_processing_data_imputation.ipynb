```python
import sys
sys.path.append('utils')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sp

plt.style.use('seaborn')

import dataframe_utils
```

In this notebook, the only necessary pre process will be the removal of outliers with over 4 std.

```python
df_import = pd.read_csv('dataset/train_rf_imputed.csv')
display(df_import.head(10))
print(df_import.shape)
```

```python
df_import.describe()
```

```python
outlier_threshold = 4
numerical_columns = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount','Base_Loan_Installment','Remaining_Income']
z_score = np.abs(sp.zscore(df_import[numerical_columns]))
df_numerical_no_outlier = df_import.loc[(z_score < outlier_threshold).all(axis=1),:].copy()
display(df_numerical_no_outlier.describe())
print(df_numerical_no_outlier.shape)
```

```python
df_numerical_no_outlier.to_csv('dataset/train_loan_rf_no_outlier.csv',index = False)
```
